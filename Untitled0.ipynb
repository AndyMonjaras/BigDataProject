```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Preparar datos (X son las columnas numéricas, quitamos Label)
X = df_pandas.drop('Label', axis=1)

# 2. Escalar datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Entrenar Isolation Forest (El Cerebro)
# Usamos 10% de contaminación (ajustable)
model = IsolationForest(contamination=0.1, random_state=42)
preds = model.fit_predict(X_scaled)

# 4. CREAR LA COLUMNA QUE FALTA (Aquí se arregla el error)
df_pandas['is_anomaly'] = np.where(preds == -1, 'Posible Ataque', 'Normal')

# 5. Generar la Evidencia Visual (Fase 4)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=df_pandas,
    x="Flow Duration",
    y="Total Length of Fwd Packets",
    hue="is_anomaly",
    palette={"Normal": "#2ecc71", "Posible Ataque": "#e74c3c"}, # Verde y Rojo
    alpha=0.5
)

plt.title("Fase 4: Detección Automática de Anomalías (IA)")
plt.xlabel("Duración del Flujo (ms) - Escala Log")
plt.ylabel("Bytes Enviados")
plt.xscale('log') # Para ver mejor la dispersión
plt.legend(title="Veredicto de la IA")
plt.show()

# 6. Ver resumen de coherencia
print("\n--- RESUMEN DE DETECCIÓN ---")
print(df_pandas.groupby(['Label', 'is_anomaly']).size().unstack(fill_value=0))
```

```python
# 1. Limpiamos espacios en blanco de TODAS las columnas primero
for col_name in df_spark.columns:
    df_spark = df_spark.withColumnRenamed(col_name, col_name.strip())

# 2. Tu lista de columnas seleccionadas (asegurándonos que los nombres coincidan)
columnas_proyecto = [
    "Destination Port",
    "Flow Duration",
    "Total Fwd Packets",
    "Total Backward Packets",
    "Total Length of Fwd Packets",
    "Total Length of Bwd Packets",
    "Fwd Packet Length Max",
    "Fwd Packet Length Min",
    "Fwd Packet Length Mean",
    "Fwd Packet Length Std",
    "Label"
]

try:
    # 3. Creamos el DataFrame final solo con lo que necesitas
    df_final = df_spark.select(*columnas_proyecto)

    # 4. Limpieza rápida: Eliminar filas con valores nulos o infinitos
    # (Muy común en este dataset de red)
    df_final = df_final.dropna()

    print("✅ Columnas seleccionadas y datos limpios.")
    df_final.show(5)

    # 5. El Puente Técnico: Pasar a Pandas (Muestra de 50k filas)
    df_pandas = df_final.limit(50000).toPandas()
    print(f"✅ ¡Listo! df_pandas creado con {df_pandas.shape[0]} filas y {df_pandas.shape[1]} columnas.")

except Exception as e:
    print(f"❌ Error en la selección: {e}")
    print("Revisa si algún nombre de columna tiene una letra distinta.")
```
